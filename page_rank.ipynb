*Some basic imports used to implement the algorithms and visualize the networks*

```python
import networkx as nx
import matplotlib.pyplot as plt
```

*Read in Adjecency List from a .txt file in a subdirectory into a nx directed graph object*

```python
def read_adjlist(filename : str):
     with open(filename, 'rb') as infile: # create a filehandler to open the sample data
         return nx.read_adjlist(infile, create_using=nx.DiGraph()) # read it into a nx directed graph object
```

```python
G = read_adjlist("PageRankExampleData/medium.txt")
fig = plt.figure(figsize=(15,10))
nx.draw(G, with_labels=True, node_color='green')
```
